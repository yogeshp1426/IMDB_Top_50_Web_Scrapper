```python
## Importing needed libraries for the project of web scrapping like BeautifulSoup from bs4 for saving and extracting html components
## requests to support different kinds of http requests to be made
## urllib is used for opening URLs so we can access data from internet like XML, HTML, JSON etc

from bs4 import BeautifulSoup as soup
import requests
from urllib.request import urlopen as uReq
import csv

# a List to be used as container
csv_rows = [] 

# defining url from which we want the html page to be scrapped
# After that we use urllib to open and accesss the url and close the connection once it is captured
# page_soup is used to convert the html page captured into useable html container using soup

my_url = ("https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012")
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html,"html.parser")

# After examining the html one can see that the the div  with class lister-item-content has each movie in a seperate binned way
movie_container = page_soup.findAll("div",{"class":"lister-item-content"})

# we open a file using f open to write out data we wrote into csv format using csv library imported at the start 
f  = open('imdb_dataout.csv','w',newline = '')
writer = csv.writer(f)

# Iterating through each movie in movie_container and accessing the movie attributes like title, runtime, certificate etc
# And writing out each of that row created into the file opened above
# Also dont forget to close the file opened above using f.close or else any changes made to file will throw error
# of 'file still in use'

for movie in movie_container:
        title = movie.h3.a.text
        stuff_container = movie.findAll("span")
        release_year = stuff_container[1].text
        certificate = stuff_container[2].text
        runtime = stuff_container[4].text
        genres = stuff_container[6].get_text(strip = True)
        rating = stuff_container[25].text
        votes = stuff_container[33].text
        gross_making = stuff_container[36].text
        csv_row = [title,release_year,certificate,runtime,genres,rating,votes,gross_making]
        print(csv_row)
        writer.writerow(csv_row)
        csv_rows.append(csv_row)
#print(csv_rows)
f.close()        
    
```
